## Rename columns

```python
df.rename(columns={'Latitude': 'Y',
                  'Longitude': 'X'}, inplace=True)
df.head()
```

## Empty values

```python
for col in df:
    print(col, df[col].isnull().sum() )
```

Closed nan values are fine, because the case is still open and we will only look at the open date. Status notes and request type are also not useful. Basically we need to take care of nan values for Location, Supervisor District, Neighborhood and Police District

```python
df[df['Supervisor District'].isnull()]
```

We see a problem. The rows with no district are the ones with location (0,0). Since we are going to work based on locations. Drop all the locations with (0,0)

```python
df = df[df.Point != '(0.0, 0.0)']
df[df['Supervisor District'].isnull()]
```

We got rid from a lot of missing location rows, but we are still missing some supervisor districts. We can get the correct district using the location later with our class.

```python
print('Nan on Neighborhood', df[df['Neighborhood'].isnull()].shape[0])
print('Nan on Police District', df[df['Police District'].isnull()].shape[0])
```

```python
df[df['Neighborhood'].isnull()]
```

There are less than 81 rows with nan out of 59895, that is 0.1%. Therefore, we will drop them without risk of losing data

```python
df.dropna(subset=['Neighborhood'], inplace=True)
df.dropna(subset=['Police District'], inplace=True)
df.dropna(subset=['Supervisor District'], inplace=True)
```

```python
print('Nan on Neighborhood', df[df['Neighborhood'].isnull()].shape[0])
print('Nan on Police District', df[df['Police District'].isnull()].shape[0])
print('Nan on Supervisor District', df[df['Supervisor District'].isnull()].shape[0])
```

## Datetime objects

```python
df['Opened'] = pd.to_datetime(df['Opened'] )
df['Closed'] = pd.to_datetime(df['Closed'] )
df['Updated'] = pd.to_datetime(df['Updated'] )
```

```python
df['Year'] = df['Opened'].dt.year
df['Month'] = df['Opened'].dt.month
df['Day'] = df['Opened'].dt.day
```

```python
df.head()
```

## Readability

```python
df['Police District'] = df['Police District'].astype(str).str.lower() 
df['Supervisor District'] = df['Supervisor District'].astype(str).str.lower() 
df['Neighborhood'] = df['Neighborhood'].astype(str).str.lower() 
df['Address'] = df['Address'].astype(str).str.lower() 
df.head()
```

```python
print("Min date: ",min(df['Opened']))
print("Max date: ",max(df['Opened']))
```
