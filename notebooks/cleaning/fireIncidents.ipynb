```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('../../data/raw/Fire_Incidents.csv')
```

## Empty values

```python
df[df['Location'].isnull()]
```

```python
df.dtypes
```

```python
df = df.dropna(subset=['Location'])
print("Rows with nan location: ", df[df['Location'].isnull()].shape[0]
```

## Datetime objects 

Convert to datetime object

```python
df['Alarm DtTm'] = pd.to_datetime(df['Alarm DtTm'] )
df['Arrival DtTm'] = pd.to_datetime(df['Arrival DtTm'] )
df['Close DtTm'] = pd.to_datetime(df['Close DtTm'] )
```

```python
df.dtypes
```

We will take the alarm dttm for the year, month and day of the eviction

```python
df['Year'] = df['Alarm DtTm'].dt.year
df['Month'] = df['Alarm DtTm'].dt.month
df['Day'] = df['Alarm DtTm'].dt.day
```

```python
df.head()
```

## Location coordinates

```python
df['Location'] = df['Location'].str.extract(r"\((.*?)\)", expand=False)
new = df["Location"].str.split(" ", n = 1, expand = True) 
new
```

```python
df["X"] = new[0] 
df["Y"] = new[1]
df.head(10)
```

## Nan values (for columns that match other datasets)

```python
for col in df:
    print(col, df[col].isnull().sum() )
```

```python
df[df['Supervisor District'].isnull()].head()
```

```python
df = df.dropna(subset=['Supervisor District'])
print("Rows with nan Supervisor District: ", df[df['Supervisor District'].isnull()].shape[0])
```

```python
df.rename(columns={'Neighborhood  District':'Neighborhood District'},
          inplace=True)
df[df['Neighborhood District'].isnull()].head()
```

```python
df = df.dropna(subset=['Neighborhood District'])
print("Rows with nan Neighborhood District: ", df[df['Neighborhood District'].isnull()].shape[0])
```

Finally for readibility between data sets, we will apply lower case to all values

```python
df = df.apply(lambda x: x.astype(str).str.lower())
df.head()
```

```python
print("Min date: ",min(df['Alarm DtTm']))
print("Max date: ",max(df['Alarm DtTm']))
```

```python
df.to_csv('../../data/processed/fireIncidents.csv')
```
