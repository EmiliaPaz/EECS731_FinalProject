```python
def getPercent(list1):
    evList=[]
    listsum=0
    for x in list1:
        listsum=listsum+x
    for x in list1:
        evList.append(x/listsum)
    return(evList)
```

```python
illegalPerc=getPercent(illegal_districts)
graffitiPerc=getPercent(graffiti_districts)
firePerc=getPercent(fire_districts)
noisePerc=getPercent(noise_districts)
homelessPerc=getPercent(homeless_districts)
evictionPerc=getPercent(eviction_districts)
dpPerc=getPercent(dp_districts)
```
